### (c) Modelo estatístico que permita a predição das vendas a partir de variações na temperatura, bem como analisa acerca da sensibilidade das vendas à tais oscilações

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

#### 1. Construção de um modelo Baseline

O modelo baseline servirá como base de comparação para outros modelos.

**Modelo escolhido:** Regressão Linear

**Atributos escolhidos:** tempMean. O fato de não escolher o temp, é que ele é muito correlacionado com a variável alvo.

```python
df.info()
```

```python
for col in ['temp','tempMean']:
    dfbase = df[[col,'vendas']].dropna().copy()
    X = np.array(dfbase[col]).reshape(-1,1)
    Y = np.array(dfbase['vendas']).reshape(-1,1)
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
    model = LinearRegression()
    model.fit(X_train,Y_train)
    print('Score obtido com a utilização da variável :{}'.format(col))
    print('Score: {}'.format(model.score(X_test,Y_test)))
```

## Conclusão prévia:

O resultado do baseline foi muito ruim, como esperado. O fato de ter sido escolhido apenas a variação da temperatura média dos períodos de vendas como variável preditora é uma forma de mostrar que correlação não necessariamente tem haver com casualidade, se puder-mos notar a correlação da variável 'temp' que foi utilizada como base para construção da temperatura média mensal, é possível notar uma correlação superior a 0.9 e isso ao ser testado no modelo, faz com que nosso modelo tenha um coeficiente de determinação elevado, ou seja, um ajuste quase perfeito com a variável dependente. Esse é um problema de Séries Temporais, e uma simples regressão linear não é um método adequado para resolução desse tipo de problema. Com esse motivo então que fiz a escolha de criar variáveis sazonais(Dia, Mês, Ano, Dia da Semana), que podem ser utilizadas no meu modelo para tentar capturar algumas características das vendas que não podem ser capturadas apenas utilizando informações pontuais.
