```python
# Import Libraries
import pandas as pd
import numpy as np
```

```python
# Open Dataset

df = pd.read_excel('/home/jovyan/workspace/data/dados_clima.xlsx')
df.head()
```

```python
# removendo colunas desnecessárias
colunasDesnec = ['Unnamed: {}'.format(num) for num in [4,9,14,19]]
df.drop(colunasDesnec, axis = 1, inplace = True)
```

```python
df.head()
```

```python
# Colunas Base
colunas = ['UF','Temperatura (celsius)','Chuva (mm)','Data']

for identificador in ['','.1','.2','.3','.4']:
    if identificador == '':
        newDf = df[colunas].copy()
    else:
        colTemp = ['{}{}'.format(col,identificador) for col in ['UF','Temperatura (celsius)','Chuva (mm)','Data']]
        dicioAux = {antigo:novo for antigo,novo in zip(colTemp,colunas)}
        dfTemp = df[colTemp].rename(dicioAux,axis='columns').copy()
        newDf = newDf.append(dfTemp)
```

```python
newDf
```

```python
# Salvando base de Dados
newDf.to_csv('/home/jovyan/workspace/data/dados_clima_corrigido.csv', index = False)
```
